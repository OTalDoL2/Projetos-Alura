```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
pd.read_csv(url)
```

```python
dados = pd.read_csv(url, sep=';')
dados
```

```python
# Método que pega as x primeiras linhas
dados.head(10)

# Método que pega as x últimas linhas
dados.tail(10)                                                 
```

```python
# Mostra uma determinada coluna
dados['Tipo']

# Mostra determinadas colunas
dados[['Quartos', 'Valor']]
```

```python
# Calcula as médias
dados.mean()

# Calcula a média
dados['Valor'].mean()

# Método de Agrupamento / Agregação
dados.groupby('Tipo').mean(numeric_only=True)

# O único valor que ele pegará a média será Valor, e o Tipo será o index
dados.groupby('Tipo')['Valor'].mean() 
```

```python
# Ele filtra do menor para o maior Valor
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

# Plota o gráfico
df_preco_tipo.plot(kind='barh', figsize=(14,10), color='purple')
```
